```python
ds = pd.read_csv(filename)
```

```python
persons = set()
```

```python
for i in range(len(ds)):
    persons.add(ds["Phone1"][i])
    persons.add(ds["Phone2"][i])
```

```python
total_person_count = len(persons)
```

```python
dates = []
```

```python
for i in range(len(ds)):
    dates.append(ds["When"][i].split(' ')[0])
```

```python
total_days_count = len(dates)
```

```python
ds["WhenDate"] = dates
```

```python
group_by_date = ds.groupby('WhenDate').size()
```

```python
min_contacts_per_day = group_by_date.min()
```

```python
max_contacts_per_day = group_by_date.max()
```

```python
average_contacts_per_day = group_by_date.mean()
```
